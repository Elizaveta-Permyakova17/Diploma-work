```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Parameters of the model
kappa = 0.1
alpha = 0.2
b1 = 0.4
b2 = 0.4
b3 = 0.4
a1 = 0.1
k1 = 0.5
g1 = 0.2
g3 = 0.2
U0 = 2
```

```python
tau = 0.01 #It's our step, "delta t"
N = 100000 #Number of iterations
```

```python
beta = 0.5
Beta = np.zeros([201])
for j in range (0, 201):
    Beta[j] = beta
    beta += 0.005
```

## a3 = 0.2

```python
#Creating arrays of zeros for variables
U = np.zeros([N])
I = np.zeros([N])
C1 = np.zeros([N])
M1 = np.zeros([N])
```

```python
#Initial conditions
U_0 = 2.0
I_0 = 0.0
M1_0 = 0.5
C1_0 = 0.08
```

```python
U[0] = U_0
I[0] = I_0
M1[0] = M1_0
C1[0] = C1_0
```

```python
#Analysis
beta = 0.5
a3 = 0.2
```

```python
#For 200 steps of changing beta
Cytokines = np.zeros([201])
Macrophages = np.zeros([201])
Uninflamed = np.zeros([201])
Inflamed = np.zeros([201])
```

```python
for i in range (0, 201):
    for j in range(1,N):
        U[j] = tau * (kappa * (U0 - U[j-1]) - (alpha * U[j-1] * C1[j-1])/(1 + b1 * C1[j-1])) + U[j-1]
        I[j] = tau * ((alpha * U[j-1] * C1[j-1])/(1 + b1 * C1[j-1]) - beta * I[j-1]) + I[j-1]
        C1[j] = tau * ((a1 * C1[j-1] * M1[j-1])/(1 + b2 * C1[j-1]) + k1 * I[j-1] - g1 * C1[j-1]) + C1[j-1]
        M1[j] = tau * ((a3 * C1[j-1])/(1 + b3 * C1[j-1]) - g3 * M1[j-1]) + M1[j-1]
    Cytokines[i] = C1[N-1] #saving the last result
    Macrophages[i] = M1[N-1]
    Uninflamed[i] = U[N-1]
    Inflamed[i] = I[N-1]
    beta += 0.005
```

```python
beta = 0.5
Beta0 = np.zeros([101])
for j in range (0, 101):
    Beta0[j] = beta
    beta += 0.005
C0 = np.zeros([101])
```

```python
plt.figure(figsize=(10,5))
#plt.title("Dependence of C1 on beta with a3 = 0.2")

plt.plot(Beta, Cytokines, c = 'blue')
plt.plot(Beta0, C0, c = 'red')

plt.xticks(np.arange(0.5, 1.6, 0.1))
plt.xlabel('$beta$')
plt.ylabel('$C1$')
plt.legend(['устойчиво', 'неустойчиво'])
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10,5))

plt.plot(Beta, Macrophages, c = 'blue')


plt.xticks(np.arange(0.5, 1.6, 0.1))
plt.xlabel('$beta$')
plt.ylabel('$M1$')
plt.legend(['устойчиво', 'неустойчиво'])
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10,5))

plt.plot(Beta, Uninflamed, c = 'blue')


plt.xticks(np.arange(0.5, 1.6, 0.1))
plt.xlabel('$beta$')
plt.ylabel('$U$')
plt.legend(['устойчиво', 'неустойчиво'])
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10,5))

plt.plot(Beta, Inflamed, c = 'blue')


plt.xticks(np.arange(0.5, 1.6, 0.1))
plt.xlabel('$beta$')
plt.ylabel('$I$')
plt.legend(['устойчиво', 'неустойчиво'])
plt.grid()
plt.show()
```

```python
Cytokines2 = np.zeros([121])
Macrophages2 = np.zeros([121])
Uninflamed2 = np.zeros([121])
Inflamed2 = np.zeros([121])
```

```python
for i in range (0, 121):
    for j in range(1,N):
        U[j] = tau * (kappa * (U_0 - U[j-1]) - (alpha * U[j-1] * C1[j-1])/(1 + b1 * C1[j-1])) + U[j-1]
        I[j] = tau * ((alpha * U[j-1] * C1[j-1])/(1 + b1 * C1[j-1]) - beta * I[j-1]) + I[j-1]
        C1[j] = tau * ((a1 * C1[j-1] * M1[j-1])/(1 + b2 * C1[j-1]) + k1 * I[j-1] - g1 * C1[j-1]) + C1[j-1]
        M1[j] = tau * ((a3 * C1[j-1])/(1 + b3 * C1[j-1]) - g3 * M1[j-1]) + M1[j-1]
    Cytokines1[i] = C1[N-1] #saving the last result
    Macrophages1[i] = M1[N-1]
    Uninflamed1[i] = U[N-1]
    Inflamed1[i] = I[N-1]
    beta += 0.005
```

```python
beta = 0.7

for i in range (0, 121):
    for j in range(1,N):
        u[j] = tau * (kappa * (u_0 - u[j-1]) - (alpha * u[j-1] * c1[j-1])/(1 + b1 * c1[j-1])) + u[j-1]
        inf[j] = tau * ((alpha * u[j-1] * c1[j-1])/(1 + b1 * c1[j-1]) - beta * inf[j-1]) + inf[j-1]
        c1[j] = tau * ((a1 * c1[j-1] * m1[j-1])/(1 + b2 * c1[j-1]) + k1 * inf[j-1] - g1 * c1[j-1]) + c1[j-1]
        m1[j] = tau * ((a3 * c1[j-1])/(1 + b3 * c1[j-1]) - g3 * m1[j-1]) + m1[j-1]
    Cytokines2[i] = c1[N-1] #saving the last result
    Macrophages2[i] = m1[N-1]
    Uninflamed2[i] = u[N-1]
    Inflamed2[i] = inf[N-1]
    beta += 0.005
```

```python
beta = 0.7
Beta0 = np.zeros([61])
for j in range (0, 61):
    Beta0[j] = beta
    beta += 0.005
C0 = np.zeros([61])
```

```python
beta = 0.7
Beta1 = []
Beta2 = []
Beta3 = []
root1 = []
root2 = []
root3 = []
while (beta < 1.501):
    a = beta * g1 * g3 * b2 * b3 * (kappa * b1 + alpha)
    b = beta * g1 * g3 * (b2 + b3) * (kappa * b1 + alpha) + kappa * g3 * b2 * b3 * (beta * g1 - alpha * k1 * U0 ) \
        - beta * a1 * a3 * (kappa * b1 + alpha)                        
    c = beta * g1 * g3 * (kappa * b1 + alpha) + kappa * g3 * (b2 + b3) * (beta * g1 - alpha * k1 * U0) - beta * a1 * a3 * kappa
    d = kappa * g3 * (beta * g1 - alpha * k1 * U0)
    A1, A2, A3 = np.roots([a, b, c, d])
    if (A1.dtype == 'float64') & (A1 > 0.):
        Beta1.append(beta)
        root1.append(A1)
    if (A2.dtype == 'float64') & (A2 > 0.):
        Beta2.append(beta)
        root2.append(A2)
    if (A3.dtype == 'float64') & (A3 > 0.):
        Beta3.append(beta)
        root3.append(A3)
    beta += 0.005
```
